```python
from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=openrouter_api_key,
)

completion = client.chat.completions.create(
  extra_headers={},
  extra_body={},
  model="qwen/qwen3-coder:free",
  messages=[
    {
      "role": "user",
      "content": "python code to compute the first 10 digits of pi"
    }
  ]
)
print(completion.choices[0].message.content)
```

```python
from decimal import Decimal, getcontext

def compute_pi_chudnovsky(precision=100):
    # Set precision for Decimal calculations
    getcontext().prec = precision
    
    def factorial(n):
        if n <= 1:
            return 1
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
    
    # Chudnovsky algorithm terms
    def chudnovsky_term(k):
        numerator = factorial(6 * k) * (545140134 * k + 13591409)
        denominator = factorial(3 * k) * (factorial(k) ** 3) * (640320 ** (3 * k))
        return Decimal(numerator) / Decimal(denominator)
    
    # Calculate sum of terms
    total = Decimal(0)
    for k in range(10):  # First few terms give high accuracy
        term = chudnovsky_term(k)
        if k % 2 == 0:
            total += term
        else:
            total -= term
    
    # Apply the constant multiplier
    pi_estimate = 1 / (12 * total)
    return pi_estimate

# Compute pi
pi_value = compute_pi_chudnovsky()
print(f"Pi: {pi_value}")
print(f"First 10 digits: {str(pi_value)[:12]}")
```
